# SN比（MT法での利用）

* 『実験計画法とタグチメソッド』参照

## SN比とは
* タグチメソッドでバラツキを見る場合に利用する。
    * タグチメソッドでは、バラツキを見る尺度は標準偏差ではなく**SN比**を用いる。

* 例：10試合の両者のスコア

|          | 鈴木 | 佐藤 |
| -------- | ---- | ---- |
| 1        | 89   | 87   |
| 2        | 91   | 88   |
| 3        | 86   | 89   |
| 4        | 88   | 90   |
| 5        | 94   | 88   |
| 6        | 86   | 89   |
| 7        | 87   | 87   |
| 8        | 84   | 87   |
| 9        | 83   | 88   |
| 10       | 82   | 87   |
| 平均     | 87.0 | 88.0 |
| 標準偏差 | 3.7  | 1.1  |

* SN比のSはSignal、信号（入力）で、品質工学における信号とは、技術の働きで蓄積される有効性だと考える。
* NはNoise、ノイズ（雑音）で、ノイズとは送られた信号に対しし、出力として有効に働かなかった有害性と考える。
* 例）
    * シャープペンシルの芯を0.05mmに設定→信号(S)＜入力＞
    * 何らかの原因で、設定通りにならない→ノイズ(N)＜出力＞
* Sは信号なので大きいほどよく、Nはノイズなので小さいほど良いと考える。
* このSとNの比とSN比といい、単位はデシベル[db]を使う。
* SN比の種類
    1. 静特性：目標値を持つ特性
    2. 動特性：入力に応じて出力値が変化する特性
* 静特性の種類：

| 特性     | 内容                         |
| -------- | ---------------------------- |
| 望大特性 | 大きい値が望ましい特性       |
| 望小特性 | 小さい値が望ましい特性       |
| 望目特性 | 目標値が近いのが望ましい特性 |

## 望大特性
* 世の中には、「品質検査合格数」「売上金額」など、値が大きいほど望ましいとされるものが存在する。それらのSN比を算出する際に用いるのが、**望大特性**のSN比
* 公式：
$$
-10 \log_{10} \left( \left( \frac{1}{D_1^2} + \dots + \frac{1}{D_n^2} \right) \cdot \frac{1}{n} \right)
$$
    * $D^2$:データの2乗(MT法では2乗のマハラノビス距離)
    * $n$:データ数

* 鈴木、佐藤のSN比を求める
    1. 各$\frac{1}{D^2}$を求める
    2. 1の合計を$n$で割る
    3. $-10log_{10}$を当てる
    * -> 鈴木さんのSN比：38.769824、佐藤さんのSN比：38.887986
* SN比は、値が大きいほどばらつきが少ない、すなわち安定していると考えるので、鈴木さんの方が安定していると言える。

## 望目特性
* 「工場で製造されるシャープペンシルの芯の直径」「ゴルフの理想スコア」など、値が「具体的なある値」に近いほど素晴らしいとされるものが存在する。すなわち、値が目指す値に近いほど望ましいとするSN比を、**望目特性**のSN比

* 公式：
$$
10 \log_{10} \left( \frac{(\overline{D})^2}{V_e} - \frac{1}{n} \right)
$$

    * $\overline{D}$:データの平均
    * $V_e$:分散
    * $n$:データ数

* 鈴木、佐藤のSN比を求める

|                    | 鈴木   | 佐藤  |
| ------------------ | ------ | ----- |
| 1                  | 89     | 87    |
| 2                  | 91     | 88    |
| 3                  | 86     | 89    |
| 4                  | 88     | 90    |
| 5                  | 94     | 88    |
| 6                  | 86     | 89    |
| 7                  | 87     | 87    |
| 8                  | 84     | 87    |
| 9                  | 83     | 88    |
| 10                 | 82     | 87    |
| 平均(Dbar)         | 87     | 88    |
| 平均の２乗(Dbar^2) | 7569   | 7744  |
| 分散V_e            | 13.556 | 1.111 |
| (Dbar^2)/V_e       | 558    | 6970  |
| (Dbar^2)/V_e - 1/n | 558    | 6970  |
| 望目特性のSN比     | 27.47  | 38.43 |

* SN比は佐藤さんが38.43で鈴木さんの27.47を上回っている
* SN比は、値が大きいほどバラツキが小さい、すなわち、安定している。よって、佐藤さんの方が鈴木さんより安定していると言える。

## 健康群のデータを用いて各マハラノビス距離を求め、判別用単位空間を作成する

```python
normal_df = data_df[data_df['健康有無'] == '健康'].drop(columns='健康有無')
abn_df = data_df[data_df['健康有無'] == '不健康'].drop(columns='健康有無')
```

```python
# 健康群データの標準化
x = normal_df.values
mean_val = np.mean(x, axis=0)
std_val = np.std(x, axis=0)
x_std = (x - mean_val) / std_val
# 分散共分散用
# x_std_df = (normal_df - mean_val) / std_val

# 相関行列の作成
corr_matrix = np.corrcoef(x_std, rowvar=False)
corr_matrix

# 分散共分散行列
# cov_matrix = x_std_df.cov()
# cov_matrix
```

* かなり相関の高い変数があるが、逆行列を求められそうなので求める

```python
# 逆行列算出
# 相関行列版
inv_corr_matrix = np.linalg.inv(corr_matrix)
inv_corr_matrix
# 分散共分散版
# inv_cov_matrix = np.linalg.inv(cov_matrix.values)
# inv_cov_matrix
```

```python
# スケーリング因子数
k = x_std.shape[1]
print(f"変量数：{k}")
```

```python
# 健康群のマハラノビス距離の算出
mahala_dists = []
for i in range(len(x_std)):
    row = x_std[i, :]
    # マハラノビス距離の算出
    D2 = (1/k) * np.dot(np.dot(row, inv_corr_matrix), row.T)
    # D2 = (1/k) * np.dot(np.dot(row, inv_cov_matrix), row.T) # 分散共分散版
    mahala_dists.append(D2) # SN比算出に使うので、2乗のまま

normal_df['mahala_dist(D^2)'] = mahala_dists
```

```python
print("健康群のマハラノビス距離")
normal_df
```

## 不健康群(異常データ)のマハラノビス距離算出

```python
# 標準化
target_rec_std = (abn_df.values - mean_val) / std_val

# マハラノビス距離の算出
new_distances = []
for i in range(len(target_rec_std)):
    row = target_rec_std[i, :]
    # マハラノビス距離の算出
    D2 = (1/k) * np.dot(np.dot(row, inv_corr_matrix), row.T)
    new_distances.append(D2)

abn_df['mahala_dist(D^2)'] = new_distances
abn_df
```

* ※上記Noごとに、単位空間を作成することになるが、Noごとに値が1の変数のみ使う

### 直交表と対象変数の設定

```python
# 直交表の作成と対象変数の設定

n1 = [1,1,1,1,2,2,2,2]
n2 = [1,1,2,2,1,1,2,2]
n3 = [1,1,2,2,2,2,1,1]
n4 = [1,2,1,2,1,2,1,2]
n5 = [1,2,1,2,2,1,2,1]
n6 = [1,2,2,1,1,2,2,1]

oa_df = pd.DataFrame({'評価1':n1,'評価2':n2,'評価3':n3,'評価4':n4,'評価5':n5,'評価6':n6}, index=[x for x in range(1,9)])

# 実験ごとの対象変数をセット
oa_df['target_cols'] = oa_df.apply(lambda row: [col for col in oa_df.columns if row[col] == 1], axis=1)
oa_df
```

### 実験Noごとにマハラノビス単位空間用の逆行列を作成

```python
# マハラノビス空間作成用関数
def make_mahala_space(data):
    """
    引数のマハラノビス空間(逆行列)を作成して返す
    ・data: pandas.dataframe マハラノビス空間を作成するデータセット
    ・return:
        ・data:レコードごとにマハラノビス距離の計算結果が含まれるdataframe
        ・mean_val: 各列の平均値
        ・std_val:各列の標準偏差
        ・inv_corr_matrix:相関行列の逆行列
    """
    x = data.values

    # 標準化
    mean_val = np.mean(x, axis=0)
    std_val = np.std(x, axis=0)
    x_std = (x - mean_val) / std_val

    if len(data.columns) > 1:
        # 相関行列の作成
        corr_matrix = np.corrcoef(x_std, rowvar=False)
        # 逆行列算出
        inv_corr_matrix = np.linalg.inv(corr_matrix)
        # 因子数抽出
        k = x_std.shape[1]
        # マハラノビス距離の算出
        mahala_dists = []
        for i in range(len(x_std)):
            row = x_std[i, :]
            D2 = (1/k) * np.dot(np.dot(row, inv_corr_matrix), row.T)
            mahala_dists.append(D2)
    else:
        # 変数1つなので、行列使わない
        mahala_dists = []
        for i in range(len(x_std)):
            D2 = np.square(x_std[i] - mean_val) / std_val
            mahala_dists.append(D2)
        inv_corr_matrix = None

    data['mahala_dist(D^2)'] = mahala_dists

    return data, mean_val, std_val, inv_corr_matrix
```

```python
# 直交表の割付けに従い、複数の単位空間を定義する

# 健康データ
ori_data = normal_df.drop(columns='mahala_dist(D^2)')

# 実験Noごとにマハラノビス単位空間作成
maha_list = []
for idx, cols in enumerate(oa_df['target_cols']):
    # 対象変数のデータ取得
    target_data = ori_data[cols].copy()
    # マハラノビス距離算出
    add_data, mean_val, std_val, inv_corr_matrix = make_mahala_space(target_data)
    maha_dict = {
        'No': idx,
        'target_cols':cols,
        'mean_val':mean_val,
        'std_val':std_val,
        'inv_corr_matrix':inv_corr_matrix}
    # 結果格納
    maha_list.append(maha_dict)
```

### 実験ごとの単位空間を用いて、不健康データのマハラノビス距離を算出

```python
# 不健康データ
ab_df = abn_df.drop(columns='mahala_dist(D^2)')
ab_df
```

```python
# マハラノビス距離算出関数
def calc_ab_m_dist(ab_record, m_dict):
    """
    ab_record: 異常レコード
    m_dict: マハラノビス距離空間の辞書
    ・return:
        ・異常レコードのマハラノビス距離
    """

    mean_val = m_dict['mean_val']
    std_val = m_dict['std_val']
    inv_corr_matrix = m_dict['inv_corr_matrix']

    # 標準化
    target_rec_std = (ab_record.values - mean_val) / std_val

    # マハラノビス距離の算出
    k = len(m_dict['target_cols'])
    if k > 1:
        D2 = (1/k) * np.dot(np.dot(target_rec_std, inv_corr_matrix), target_rec_std.T)
    else:
        # 変数1つのみ
        D2 = np.square((target_rec_std - mean_val) / std_val)[0]

    return D2
```

```python
# 不健康レコード1行ごとに、直交表8パターンのマハラノビス距離を算出する
df_list = []
for idx, ab_row in ab_df.iterrows():
    m_dict_list = []
    for m_dict in maha_list:
        # 実験対象変数の取得
        cols = m_dict["target_cols"]
        # 対象変数のレコード
        target_ab = ab_row[cols]

        # マハラノビス距離算出
        md = calc_ab_m_dist(target_ab, m_dict)
        m_dict_list.append(md)

    df = pd.DataFrame({idx:m_dict_list})
    df_list.append(df)

ab_m_dis_df = pd.concat(df_list, axis=1)
ab_m_dis_df.index = ["No"+str(x) for x in range(1,9)]
ab_m_dis_df
```

### 得られたマハラノビス距離から異常レコードごとのSN比を求める

```python
# 実験ごとのSN比算出
def calc_sn_ratio(row):
    in_values = 1/row['P'] + 1/row['Q'] + 1/row['R'] + 1/row['S'] + 1/row['T']
    sn = -10 * np.log10(in_values / 5)
    return sn

ab_m_dis_df['sn_ratio'] = ab_m_dis_df.apply(calc_sn_ratio, axis=1)
ab_m_dis_df
```

### SN比の差を求め、変数を選択する
#### 「SN比の差」について
* 「SN比の差」の値がプラスなら、その変数を用いた方がマハラノビス距離のSN比が大きくなる。つまり、マハラノビス距離の「散らばりの程度」が小さくなる。
* 「SN比の差」の値が0に近ければ、その変数を用いても用いなくてもマハラノビス距離のSN比は、対して変化しない。つまり、その変数はあってもなくても良い。
* 「SN比の差」の値がマイナスならば、その変数を用いないよりも用いた方がマハラノビス距離のSN比が小さくなる。つまり、マハラノビス距離の「散らばりの程度」が大きくなる。
* MT法では、「SN比の差」の値がプラスであるほど、その変数を用いないよりも用いた方がマハラノビス距離の「散らばりの程度」が小さくなり、有用な変数と解釈する。

| SN比の差 | -   | 変数を用いると   | 散らばり   | -   | MT法での変数   |
| -------- | --- | ---------------- | ---------- | --- | -------------- |
| +の値    | →  | SN比が大きくなる | 小さくなる | →  | 有用           |
| 0に近い  | →  | 変化しない       | 変化しない | →  | どちらでもない |
| -の値    | →  | SN比が小さくなる | 大きくなる | →  | 使わない       |

```python
# SN比の差を算出する関数
def make_factor_diff(col_name, sn_df, oa_array):
    '''
    col_name: 対象変数名
    sn_df: SN比のdataframe
    oa_array: 対象変数の直交表のnumpy.array
    ・return:
        ・以下をインデックスとしたdataframe
            ・sn_used: 使った場合のSN比の平均
            ・sn_not_used: 使わなかった場合のSN比の平均
            ・sn_diff: 使った場合のSN比の平均 - 使わなかった場合のSN比
            ・列名は、対象変数名
    '''

    # その変数を使用した実験番号一覧
    sn_used = sn_df.iloc[(oa_array==1).values]
    # その変数を使わない実験番号一覧
    sn_notused = sn_df.iloc[(oa_array==2).values]
    # 各平均
    used_mean = sn_used.mean()
    notused_mean = sn_notused.mean()
    # 差
    sn_diff = used_mean - notused_mean

    # 結果
    result_df = pd.DataFrame(
        {col_name:[used_mean, notused_mean, sn_diff]}, index=['sn_used', 'sn_not_used', 'sn_diff'])

    return result_df
```

```python
# 変数ごとに、使用した場合としない場合のSN比とその差を算出
diff_list = []
for idx, row in oa_df.drop(columns="target_cols").T.iterrows():
    result_df = make_factor_diff(idx, sn_df, row)
    diff_list.append(result_df)

sn_diff_df = pd.concat(diff_list, axis=1)
sn_diff_df
```

### 要因効果図の作成による判別
* 上記のようにSN比の差を使う以外に、グラフにして判断する方法もある

```python
# plot用にdfを変換
def make_df_for_plot(sn_diff_df):
    # 列名作成
    level1_cols = [col for col in sn_diff_df.columns for _ in range(2)]
    level2_cols = ['one', 'two'] * len(sn_diff_df.columns)
    arrays = [level1_cols, level2_cols]
    columns = pd.MultiIndex.from_arrays(arrays, names=['factor', 'level'])
    plot_df = pd.DataFrame(
        columns=columns,
        data=sn_diff_df.drop(index='sn_diff').values.T.reshape(1, -1)
    )
    return plot_df

plot_df = make_df_for_plot(sn_diff_df)
plot_df
```

```python
# 要因効果図プロット用関数
def plot_factor_effect(plot_df):

    factors = plot_df.columns.get_level_values('factor').unique()
    num_factors = len(factors)

    # 横にグラフを並べる (1行それぞれプロット)
    fig, axes = plt.subplots(1, num_factors, figsize=(16, 4), sharey=True)  # 1行N列でサブプロット作成

    # colors = ['b', 'g', 'r', 'orange']

    for i, factor in enumerate(factors):
        # "one" と "two" のデータを抽出
        y_values = plot_df[factor].values.flatten()
        x_values = [1, 2]  # one -> x=1, two -> x=2

        # サブプロットに描画
        ax = axes[i]
        ax.plot(
            x_values,
            y_values,
            marker='o',
            label=factor,
            # color=colors[i]
        )
        ax.set_title(factor)  # 各サブプロットにタイトル
        ax.set_xticks([1, 2])
        ax.set_xticklabels(['one', 'two'])
        ax.axhline(0, color='black', linewidth=0.5, linestyle='--')  # 基準線
        if i == 0:
            ax.set_ylabel('SN ratio')  # 最初のグラフだけ yラベル

    # 全体の調整
    plt.suptitle('SN ratio change')
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)  # タイトルとグラフの間隔を調整
    plt.show()

# plot
plot_factor_effect(plot_df)
```

    * one(水準1)の方が大きい場合、その変数がマハラノビス距離の増加に関与している。-> その変数が異常値の原因
